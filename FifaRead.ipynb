```python
# Lets checkout the correlation between all the categories and Ability, the base score of each player
# roster.corr()['Ability']
```

```python

# then lets throw it in an easy to use function
def getCorr(roster,position=None,to='Ability'):
    """ Function to return corelation of roster at specific position to ability"""
    if position: 
        return roster[roster['Position']==position].corr()[to]
    return roster.corr[to]
```

```python
# Lets see if we can improve on that by focusing on a single position, and only looking at the top 10 correlations
getCorr(roster,'ST').nlargest(15)
```

```python
# We can see Finishing,React, Ballctrl are some of the most influential categories. Makes sense!
# Lets checkout a different position group,
getCorr(roster,'LB').nlargest(15)
```

```python
# lets define the three basic position groups + goalies
defender_roster = roster[(roster['Position']=='LB')|(roster['Position']=='CB')|(roster['Position']=='RB')|(roster['Position']=='RWB')|(roster['Position']=='LWB')]
attacking_roster = roster[(roster['Position']=='CF')|(roster['Position']=='RW')|(roster['Position']=='LW')|(roster['Position']=='ST')]
midfield_roster = roster[ (roster['Position']=='CAM') | (roster['Position']=='CDM') | (roster['Position']=='LM') | (roster['Position']=='RM')|(roster['Position']=='CM')]
goalkeeper_roster = roster[ (roster['Position']=='GK')]
```

```python
midcor = midfield_roster.corr()['Ability']
attcor = attacking_roster.corr()['Ability']
defcor = defender_roster.corr()['Ability']
```

```python
# Lets check out the difference in correlations between position groups
group_correlations = pd.concat([midcor,attcor,defcor],axis = 1,join = 'inner')
group_correlations.columns = ['Attackers','Midfielders','Defenders']
print(group_correlations)
```
